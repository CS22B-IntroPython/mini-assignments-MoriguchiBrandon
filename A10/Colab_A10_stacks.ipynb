## Question 1: Write a function to sort a stack.

```python
def sort_stack(stack):
    # Create an temp stack
    temp_stack = []

    # loop while there are elements in the original stack
    while stack:
        # Pop an element from the original stack
        current = stack.pop()

        # Move elements from temp_stack to stack if they are larger than the current element
        while temp_stack and temp_stack[-1] < current:
            stack.append(temp_stack.pop())

        # Push the current element onto temp_stack
        temp_stack.append(current)

    # Transfer elements back to the original stack to maintain the sorted order
    while temp_stack:
        stack.append(temp_stack.pop())

    return stack


stack = [34, 3, 31, 98, 92, 23]
print(sort_stack(stack))
```

#### Question 3: Nodes

<img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQW8-twGk0hCILE-muHBnG93hZJh2Qvgs7pVA&s">

1. Use a data structure of your choice to denote the different nodes.

2. Create a function that sums all of the "children". Test with the node 10 (Expected answer is 37).


```python
#Your Answer:
graph = {
    8 : [3 , 10],
    3 : [1 , 6],
    6 : [4 , 7],
    10 : [14],
    14 : [13]
}

def sum_children(graph, node):
    sum = 0
    if node in graph:
        for child in graph[node]:
          sum += child
          sum += sum_children(graph, child)
    return sum

print(sum_children(graph, 10))
```
